```python
import sys
import pandas as pd
import csv
import numpy as np
import pandas_profiling
```

```python
df = pd.read_csv('../datasets/fatalFinal.csv')
names = df.columns
df = df.drop(names[0], axis=1)
#df["ageRangeNumerical"] = 0
#df["Month"] = 0
```

```python
for index, row in df.iterrows():
    age = row[5]
    if age < 20 and age >= 0:
        df.loc[index,"ageRangeNumerical"] = 0
    elif age >= 20 and age < 30:
        df.loc[index,"ageRangeNumerical"] = 1
    elif age >= 30 and age < 40:
        df.loc[index,"ageRangeNumerical"] = 2
    elif age >= 40 and age < 50:
        df.loc[index,"ageRangeNumerical"] = 3
    elif age >= 50 and age < 60:
        df.loc[index,"ageRangeNumerical"] = 4
    elif age >= 60 and age < 99:
        df.loc[index,"ageRangeNumerical"] = 5
        
```

```python
for index, row in df.iterrows():
    boolean = row[10]
    if boolean == True:
        df.loc[index,"mentalNumerical"] = 1
    else:
        df.loc[index,"mentalNumerical"] = 0
    
```

```python
for index, row in df.iterrows():
    s = row[2].split("-")[1]
    new = int(s)
    df.loc[index,"Month"] = new
    #if boolean == True:
    #df.loc[index,"mentalNumerical"] = 1
    
```
